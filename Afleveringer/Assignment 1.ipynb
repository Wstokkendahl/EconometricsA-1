```python
import pandas as pd
df = pd.read_stata('A1_kommune.dta')
display(df)
```

```python
vars = ['taxrev','taxrate','pop']
df[vars].describe()
```

**Opgave 2**

```python
import numpy as np
import statsmodels.api as sm
df["logtaxrev"]=np.log(df.taxrev)
df["const"]=1
model_sm = sm.OLS(df['logtaxrev'], df[['const','taxrate']]).fit()
#print(model_sm.summary())
print(model_sm.summary())
```

```python
df["logpop"]=np.log(df['pop'])
import numpy as np
import statsmodels.api as sm
model_sm2 = sm.OLS(df['logtaxrev'], df[['const','taxrate','logpop']]).fit()
#print(model_sm.summary())
print(model_sm2.summary())
```
